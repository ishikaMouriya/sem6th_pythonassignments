```python
import re

class Time:
    def __init__(self, time_str):
        self.time_str = time_str
    
    def is_valid_time(self):
        """Validates if the time string is in HH:MM:SS 24-hour format."""
        pattern = r"^(?:[01]\d|2[0-3]):[0-5]\d:[0-5]\d$"
        return bool(re.match(pattern, self.time_str))
    
    def convert_to_12_hour_format(self):
        """Converts the time from 24-hour format to 12-hour format with AM/PM."""
        if not self.is_valid_time():
            return "Invalid time format. Please enter time in HH:MM:SS format."
        
        hours, minutes, seconds = map(int, self.time_str.split(':'))
        period = "AM"
        
        if hours == 0:
            hours = 12  # Midnight case
        elif hours == 12:
            period = "PM"  # Noon case
        elif hours > 12:
            hours -= 12
            period = "PM"
        
        return f"{hours:02}:{minutes:02}:{seconds:02} {period}"

# Example usage:
time_input = input("Enter time in HH:MM:SS format (24-hour): ")
time_obj = Time(time_input)
print(time_obj.convert_to_12_hour_format())
```

```python
#q8
class Vehicle:
    def __init__(self,make,model):
        self.make=make
        self.model=model
    def display_info(self):
        print( f"the make:{self.make} and model:{self.model}")

class Car(Vehicle):
    def __init__(self,make,model,numdoors):
      super().__init__(make,model)
      self.numdoors=numdoors

    def display_info(self):
        print(f"numdoors:{self.numdoors}",end=" ")
        super().display_info()
        

v1=Vehicle(1989,"RollsRoyce")
v1.display_info()
c1=Car(1989,"Hummer",4)
c1.display_info()    
```

```python
#q9
import math
class Shape:
    def area(self):
        print("Base class area")
class Circle(Shape):
    def __init__(self,r):
        self.r=r
    def area(self):
        print(f"the area of circle is:{math.pi*self.r*self.r}")

class Rectangle(Shape):
    def __init__(self,l,b):
        self.l=l
        self.b=b
    def area(self):
        print(f"the area of rectangle is:{self.l*self.b}")

s1=Shape()
s1.area()
c1=Circle(4)
c1.area()
r1=Rectangle(2,5)
r1.area()
```

```python
#q13
class InsufficientFundsError(Exception):

    def __init__(self,balance,withdraw_amount):
        self.balance=balance
        self.withdraw_amount=withdraw_amount

    def __str__(self):
        return f"Transaction failed!Available balance:{self.balance}, Requested withdrawal: â‚¹{self.withdraw_amount}."
    
class BankAccount:
    def __init__(self,balance):
        self.balance=balance
    
    def withdraw(self,amount):
        if(amount>self.balance):
            raise InsufficientFundsError(self.balance,amount)
        self.balance-=amount
        return f"Withdrawal successful! New balance: â‚¹{self.balance}"


try:
    account=BankAccount(1000)
    print(account.withdraw(1500))
except InsufficientFundsError as e:
    print(e)
```

```python
#q14
import random
from dataclasses import dataclass

# Step 1: Define the Card Data Class
@dataclass
class Card:
    rank: str
    suit: str

    def __str__(self):
        return f"{self.rank} of {self.suit}"

# Step 2: Define the Deck Class
class Deck:
    suits = ["Hearts", "Diamonds", "Clubs", "Spades"]
    ranks = ["2", "3", "4", "5", "6", "7", "8", "9", "10", "Jack", "Queen", "King", "Ace"]

    def __init__(self):
        self.cards = [Card(rank, suit) for suit in self.suits for rank in self.ranks]
        self.shuffle()  # Shuffle deck on initialization

    def shuffle(self):
        """Shuffle the deck randomly."""
        random.shuffle(self.cards)

    def deal(self, num_cards):
        """Deal a given number of cards from the deck."""
        if num_cards > len(self.cards):
            raise ValueError("Not enough cards left in the deck!")
        hand = self.cards[:num_cards]  # Take first `num_cards` cards
        self.cards = self.cards[num_cards:]  # Remove dealt cards from deck
        return hand

# Step 3: Simulate dealing 5 cards to a player
deck = Deck()
player_hand = deck.deal(5)

# Step 4: Display the playerâ€™s hand and remaining deck size
print("Player's Hand:")
for card in player_hand:
    print(card)

print(f"\nRemaining cards in the deck: {len(deck.cards)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lambda function to convert Celsius to Kelvin
celsius_to_kelvin = lambda c: c + 273.15

# Sample temperature data (Celsius)
celsius_values = [-20, -10, 0, 10, 20, 30, 40, 50]

# Convert temperatures using lambda
kelvin_values = list(map(celsius_to_kelvin, celsius_values))

# Create DataFrame
df = pd.DataFrame({'Celsius': celsius_values, 'Kelvin': kelvin_values})

# Print DataFrame
print(df)

# Plot the data
plt.plot(df['Celsius'], df['Kelvin'], marker='o', linestyle='-', color='b', label='C to K')
plt.xlabel('Temperature in Celsius')
plt.ylabel('Temperature in Kelvin')
plt.title('Celsius to Kelvin Conversion')
plt.legend()
plt.grid(True)
plt.show()
```
